```python
## Let's define a function that calculates gaussian and returns the value

def gaussNormal(x, mu, variance):
    ## x-->denotes any data point
    ## mu--> mean, ('a' in the question)
    ## variance--> ('b' in question) 
    d1 = np.sqrt(2*np.pi*variance)
    num = (x-mu)**2
    
    arg_exp = (-1*num)/(2*variance)
    
    value = np.exp(arg_exp)/d1
    
    return value
    
```

```python
## Let's create some points
X = np.linspace(-20, 20, 100)
```

```python
## corresponding arrays to store p(x|w1) and p(x|w2)

pxw1 =  [gaussNormal(x, 2, 1) for x in X]
pxw2 =  [gaussNormal(x, 5, 1) for x in X]
```

```python
size = (10,5)
plt.figure(figsize=size, dpi=80)
plt.plot(X, pxw1, 'r')

plt.title("P(x|w1) vs x")
```

## Cauchy Distribution : ${P(x|\omega_1) = \frac{1}{\pi b} \frac{1}{1+(\frac{x-a_1}{b})^2}}$

```python
## Let's create a generic function that 
## evaluates P(x|wi) for given x, ai and b

def cond_prob(x, a, b=1):
    ## the default parameter b has been set to 1
    ## as mentioned in the question
    
    part1 = 1/(b * np.pi)
    
    den = ((x-a)/b)**2 + 1
    
    part2 = 1/den
    
    result = part1 * part2
    
    return result    
```

```python
## To find p(w1|x), we need p(x|w1) and p(x|w2) 

## First sample a list of values for x axis
## We will store linearly spaced values between -20 and 20

X= np.linspace(-40, 40, 100)
print(X)
```

For the sake of simplicity, Let's assume a variable `pagb` denotes probability of a given b, i.e, p(a|b) 

```python
def pw1gx(x):
    var1 = cond_prob(x, 3) ##a1 =3 given in the question
    var2 = var1 + cond_prob(x, 5)##a2 = 5 given in the question
    
    return var1/var2
```

```python
## Let's store some Y values for every X values
## using our old friend List Comprehension

Y = [pw1gx(x_val) for x_val in X]
```

```python
## Let's  plot it now

size = (10, 5)
plt.figure(figsize=size, dpi=86)

plt.plot(X, Y, 'r')
plt.title('P(w1|x) vs x')
```
